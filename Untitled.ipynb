```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("apps.csv")

df.head()
```

```python
# Shape of dataset
df.shape
```

```python
# Dataset information
df.info()
```

```python
    # Column names
df.columns
```

```python
df.isnull().sum()
```

```python
# Fill numeric missing values with median
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill categorical missing values with mode
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Verify
df.isnull().sum()
```

```python
# Check duplicates
df.duplicated().sum()
```

```python
# Standardize column names FIRST
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Recalculate categorical columns AFTER renaming
cat_cols = df.select_dtypes(include=['object']).columns

# Standardize text data
for col in cat_cols:
    df[col] = df[col].astype(str).str.strip().str.title()

df.head()

```

```python
df = df.drop(columns=['unnamed:_0'], errors='ignore')
```

```python
plt.figure()
plt.boxplot(df[num_cols[0]])
plt.title(f"Outlier Detection: {num_cols[0]}")
plt.show()
```

```python

# Rename columns
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Drop index column
df = df.drop(columns=['unnamed:_0'], errors='ignore')

# Identify columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns
```

```python
# Boxplot for detecting outliers
plt.figure()
df[num_cols[0]].plot(kind='box')
plt.title("Outlier Detection")
plt.show()
```

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    df = df[(df[col] >= lower) & (df[col] <= upper)]

df.shape
```

```python
print("Final Shape:", df.shape)
print("\nMissing Values:\n", df.isnull().sum())
```

```python
df.to_csv("apps_cleaned.csv", index=False)
```
